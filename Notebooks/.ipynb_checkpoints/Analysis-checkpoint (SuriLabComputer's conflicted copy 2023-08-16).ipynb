```julia
function UnpackArrayList(arrayList)
    array = []
    for n in 1:length(arrayList)
        array = vcat(array, arrayList[n])
    end 
    return array
end

function RepeatList(list, n)
    retList = []
    for t in 1:length(list)
        for k in 1:n
            push!(retList, list[t])
        end 
    end 
    return retList
end
    
```
