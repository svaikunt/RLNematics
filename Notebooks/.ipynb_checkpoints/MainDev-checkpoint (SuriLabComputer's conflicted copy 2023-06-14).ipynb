```julia
# using Profile
# using ProfileView
using BenchmarkTools
```

```julia
## create the simParams struct
Nx = 50
Ny = 50
ndt = 20
bc = "pbc"
lambdaBE = 0.7
A0BE = 0.1
UBE = 3.5 
LBE = 0.1
GammaBE = 0.13
friction = 10

initRandBool = false
offX = 15
offY = 15
q = 0

cutoff = 7.5
width = 2

sP = SimParams(Nx, Ny, ndt, bc, bc, lambdaBE, A0BE, UBE, LBE, GammaBE, friction, initRandBool, [offX, offY, q], cutoff, width)

## create the forceLawParams struct
ks = 5
l0 = 15
omega = 0
mup = 1
mut = 1

fLP = ForceLawParams(ks, l0, omega, mup, mut)

# define any keyword arguments
nSteps = 100
bounds = 1e-3

env = NematicEnv(sP, fLP; nSteps = nSteps, bounds = bounds)
```

```julia
nEpisodes = 50
ex = CreateNematicExperiment(env; nEpisodes = nEpisodes)
```

```julia
run(ex)
```
